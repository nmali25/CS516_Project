```python
def clean_dti(value):
    if isinstance(value, str):
        value = value.strip()
        if value == "<20%":
            return 10
        elif value == "20%-<30%":
            return 25
        elif value == "30%-<36%":
            return 33
        elif value == "50%-60%":
            return 55
        elif value == ">60%":
            return 65
        elif value.isdigit():
            return int(value)
        else:
            try:
                return float(value)
            except:
                return None
    elif isinstance(value, (int, float)):
        return value
    return None

# Apply the function
data["dti_clean"] = data["debt_to_income_ratio"].apply(clean_dti)
```
